```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python

instagram_hours = np.array([1, 2, 3, 4, 5, 6, 7])
battery_percentage = np.array([85, 80, 75, 70, 65, 60, 55])

x = instagram_hours.reshape(-1, 1)
y = battery_percentage
```

```python
plt.scatter(instagram_hours, battery_percentage)
plt.xlabel("Instagram Usage Hours")
plt.ylabel("Battery Percentage")
plt.title("Instagram Usage vs Battery Percentage")
plt.show()
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
pred_battery = model.predict(x)
print(pred_battery)
```

```python
plt.scatter(instagram_hours, battery_percentage)
plt.plot(instagram_hours, pred_battery)

plt.show()
```

```python
mse = mean_squared_error(battery_percentage, pred_battery)
print(mse)
```

a lower MSE indicates that the model predictions are closer to the actual battery percentages
