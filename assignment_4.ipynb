QUESTION 1

```python
df = pd.read_csv("/content/ipl_player_stats.csv")

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_columns = ['runs','wickets','matches','strike_rate']
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

print(df)

df.to_csv("ipl_scaled.csv",index=False)
```

```python
df = pd.read_csv("/content/flipkart_products.csv")

from sklearn.preprocessing import MinMaxScaler
print(" before scaling")
print(df.min())
print(df.max())


scaler = MinMaxScaler()
df[:] = scaler.fit_transform(df)

print(" after scaling")
print(df.min())
print(df.max())

df.to_csv("flipkart_scaled.csv",index=False)
```

```python
df = pd.read_csv("/content/spotify_songs.csv")

corr = df.corr()
pop_corr = corr['popularity']

top_features = pop_corr.drop('popularity').sort_values(ascending=False).head(2)

print(top_features)
top_features.to_csv("top_features.csv")
```

QUESTION 4

```python
df = pd.read_csv("/content/zomato.csv")

from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor


X = df.drop("average_cost_for_two", axis=1)
y = df["average_cost_for_two"]

model = DecisionTreeRegressor()

rfe = RFE(
    estimator=model,
    n_features_to_select=3
)

rfe.fit(X, y)

print("Selected Features:")
print(X.columns[rfe.support_])

print("Ranking:")
print(rfe.ranking_)
```

```python
df = pd.read_csv("/content/bookmyshow.csv")


Q1 = df['user_ratings'].quantile(0.25)
Q3 = df['user_ratings'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

new_df = df[(df['user_ratings'] >= lower) &
            (df['user_ratings'] <= upper)]

print("Rows dropped =", len(df) - len(new_df))
```
