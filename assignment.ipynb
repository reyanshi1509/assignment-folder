```python
# 1) display first 5 rows ?

print(df.head())
```

```python
# 2) display last 3 rows ?

print(df.tail(3))
```

```python
# how to check dataset information ?

print(df.info())
```

```python
# statistical summary of dataset ?

print(df.describe())
```

```python
# check missing values ?

print(df.isnull())
```

```python
# count missing values ?

print(df.isnull().sum())
```

```python
# fill missing values using mean ?

df["Price"] = df["Price"].fillna(df["Price"].mean())

print(df)
```

```python
# fill missing vlaues using mode ?

df["Fuel_Type"] = df["Fuel_Type"].fillna(df["Fuel_Type"].mode()[0])

print(df)
```

```python
# drop missing vlaues ?

df = df.dropna()
print(df)
```

```python
# display specific column ?

print(df["Brand"])
```

```python
# access rows from 10th to 15th ?

print(df.iloc[9:15])
```

```python
# sort dataset using price column ?

print(df.sort_values(by="Price"))
```

```python
# sort dataset in descending order ?

print(df.sort_values(by="Price", ascending=False))
```

```python
# reset index of dataset ?

df = df.reset_index(drop=True)

print(df)
```

```python
# filtering using query ?

print(df.query("Price > 2000000"))
```

```python
# detect outliers using z score ?

from scipy import stats
import numpy as np

z_score = np.abs(stats.zscore(df["Price"], nan_policy='omit'))
print(z_score)
```

```python
# save cleaned dataset into new csv file ?

df.to_csv("cleaned_car_data.csv", index=False)
print("File Saved Successfully")
```
