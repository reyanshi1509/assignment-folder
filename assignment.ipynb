```python
# 1. Sum of first n positive integers

n = int(input("Enter a number: "))

sum_n = n * (n + 1) // 2

print("Sum =", sum_n)
```

```python
# 2. Count occurrences of a substring in a string

string = input("Enter main string: ")
substring = input("Enter substring: ")

count = string.count(substring)

print("Occurrences:", count)
```

```python
# 3. Count occurrences of each word in a sentence

sentence = input("Enter a sentence: ")

words = sentence.split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```

```python
# 4. Swap first two characters of two strings

str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

new_str1 = str2[:2] + str1[2:]
new_str2 = str1[:2] + str2[2:]

result = new_str1 + " " + new_str2

print(result)
```

```python
# 5. Add 'ing' or 'ly' to a string

string = input("Enter a string: ")

if len(string) < 3:
    print(string)
else:
    if string.endswith("ing"):
        print(string + "ly")
    else:
        print(string + "ing")
```

```python
# 6. Replace 'not...poor' with 'good' in a string

s = input("Enter sentence: ")

not_pos = s.find("not")
poor_pos = s.find("poor")

if not_pos != -1 and poor_pos != -1 and not_pos < poor_pos:
    s = s[:not_pos] + "good" + s[poor_pos+4:]

print(s)
```

```python
# 7. Find GCD of two numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    a, b = b, a % b

print(a)
```

```python
# 8. Check if a list contains a sublist

main_list = [1, 2, 3, 4, 5]
sub_list = [3, 4]

if str(sub_list)[1:-1] in str(main_list):
    print("Yes")
else:
    print("No")
```

```python
# 13. Sort a dictionary by value (ascending / descending)

data = {'a': 3, 'b': 1, 'c': 2}

# Ascending
asc = dict(sorted(data.items(), key=lambda x: x[1]))
print("Ascending:", asc)

# Descending
desc = dict(sorted(data.items(), key=lambda x: x[1], reverse=True))
print("Descending:", desc)
```

```python
# 14. Find the highest 3 values in a dictionary

data = {'a': 10, 'b': 5, 'c': 20, 'd': 15}

values = sorted(data.values(), reverse=True)

print("Top 3 values:", values[:3])
```

```python
# 15. Fibonacci series up to n terms 

n = 7

a, b = 0, 1

print("First few Fibonacci numbers are:")

for i in range(n + 1):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# 16. Count frequencies in a list using dictionary

data = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1, 4, 4, 4, 2, 2, 2, 2]

freq = {}

for i in data:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

for k in sorted(freq):
    print(k, ":", freq[k], end=" , ")
```

```python
#  17. Sum of odd and even series using function

def fact(n):
    f = 1
    for i in range(1, n+1):
        f = f * i
    return f

def series(n):
    odd = 9
    even = 10

    for i in range(1, n+1):
        if i % 2 == 0:
            even = even + (i*i)/fact(i)
        else:
            odd = odd + (i*i)/fact(i)

    print("Odd sum =", odd)
    print("Even sum =", even)

series(6)
```

```python
# 18. Factorial using recursion

def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)

print(fact(5))
```

```python
# 19. Return list with unique elements

def unique(lst):
    new = []
    for i in lst:
        if i not in new:
            new.append(i)
    return new

print(unique([1,2,2,3,4,4]))
```

```python
# 20. Mini Project: Password Generator

import random

class User:
    def __init__(self, id, name, password):
        self.data = (id, name, password)

def password(text):
    try:
        word = random.choice(text.split())
        pwd = word.upper() + "12@"   

        if len(pwd) < 8:
            pwd = pwd + "abc"

        return pwd
    except:
        print("Error")

name = input("Name: ")
id = int(input("ID: "))
text = input("Enter words: ")

pwd = password(text)

u = User(id, name, pwd)

print(u.data)
```
