2. You add a restaurant_id column to the restaurants table as its primary key. What is a primary key, and why must its values be unique?

Answer:
A Primary Key is a column that uniquely identifies each record in a table.
Its values must be unique and cannot contain NULL values.

Example: 'restaurant_id'

 3. The ratings table references restaurant_id from the restaurants table. What is a foreign key, and what integrity rule does it enforce?

Answer:
A Foreign Key is a column used to connect two tables.
It enforces referential integrity by ensuring that referenced values exist in the parent table.

Example: 'restaurant_id' in ratings table references restaurants table.

 4. You apply a NOT NULL constraint to the restaurant_name column. What kind of data entry does this constraint prevent?

Answer:
The NOT NULL constraint prevents empty or missing values from being inserted into a column.

Example: Restaurant name cannot be left blank.

 5. You draw an ER diagram before writing any SQL. What does an ER diagram represent in relational database design?

Answer:
An ER (Entity Relationship) Diagram represents entities, attributes, and relationships between tables in a relational database.
