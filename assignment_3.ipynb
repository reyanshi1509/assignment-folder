QUESTION 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/ipl_sample_data.csv")
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

QUESTION 2

```python
# replace missing values in player_age with the median

median_age = df['player_age'].median()
df['player_age'].fillna(median_age)
```

QUESTION 3

```python
# replace missing team names with " unknown "

df['team'] = df['team'].fillna('Unknown')
print(df.head(10))
```

QUESTION 4

```python
# one-hot encoding of venue

venue_encoded = pd.get_dummies(df['venue'])
df_encoded = pd.concat([df,venue_encoded],axis=1)

print(df_encoded)
```

QUESTION 5

```python
# encode player_role using LabelEncoder

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['player_role_encoded']= label_encoder.fit_transform(df['player_role'])

mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(mapping)

print(df[['player_role','player_role_encoded']])
```
